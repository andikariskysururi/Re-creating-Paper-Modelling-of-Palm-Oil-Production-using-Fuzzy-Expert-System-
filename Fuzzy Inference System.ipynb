##Aggregating Rule

```python
rule_k_set = [rule1_k, rule2_k, rule3_k, rule4_k, rule5_k, rule6_k, rule7_k, rule8_k, rule9_k,
               rule10_k, rule11_k, rule12_k, rule13_k, rule14_k, rule15_k, rule16_k, rule17_k,
               rule18_k, rule19_k, rule20_k, rule21_k, rule22_k, rule23_k, rule24_k, rule25_k,
               rule26_k, rule27_k]

aggregated_rule_k = np.zeros_like(KW16)

for i in rule_k_set:
    aggregated_rule_k = np.fmax(aggregated_rule_k, i)
```

##Defuzzifikasi

```python
# Defuzifikasi
KW16_defuzz = fuzz.defuzz(KW16, aggregated_rule_k, 'centroid')
KW16_activation = fuzz.interp_membership(KW16, aggregated_rule_k, KW16_defuzz)
KK6_defuzz = fuzz.defuzz(KK6, aggregated_rule_k, 'centroid')
KK6_activation = fuzz.interp_membership(KK6, aggregated_rule_k, KK6_defuzz)
print('Center of Gravity KW16:', KW16_defuzz, "% of kernel loss in LTDS I&II (KW16)")
print('Derajat Keanggotaan KW16  :', KW16_activation, '\n')
print('Center of Gravity KK6 :', KK6_defuzz," % of dirt content in palm kernel (KK6)")
print('Derajat Keanggotaan KK6 :', KK6_activation, '\n')
```
